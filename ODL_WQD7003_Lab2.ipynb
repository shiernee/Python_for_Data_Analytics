# Data Processing Using Arrays

Using NumPy arrays enables you to express many kinds of data processing tasks as concise array expressions that might otherwise require writing loops. This practice of replacing explicit loops with array expressions is commonly referred to as vectorization.

In general, vectorized array operations will often be one or two (or more) orders of magnitude faster than their pure Python equivalents, with the biggest impact in any kind of numerical computations.

```python
import numpy as np

points = np.arange(-5, 5, 0.01) # 1000 equally spaced points

# The np.meshgrid function takes two 1D arrays and produces
# two 2D matrices corresponding to all pairs of (x, y) in the two arrays:

xs, ys = np.meshgrid(points, points)

print(xs)
print()
print(ys)
```

```python
# used the matplotlib function imshow to create an image plot from a 2D array of function values

import matplotlib.pyplot as plt
z = np.sqrt(xs ** 2 + ys ** 2)

plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
```

## Expressing Conditional Logic as Array Operations

The numpy.where function return elements chosen from x or y depending on condition

```python
arr = np.array([-2.57082745,  0.02373729,  1.24391309, -1.50603771])
arr
```

```python
np.where(arr > 0)
```

```python
# return 2 if satisfy the condition, else return -2
np.where(arr > 0, 2, -2)
```

```python
# return 2 if satisfy the condition, else return the original value

np.where(arr > 0, 2, arr)
```

## Sorting

```python
arr = np.random.randn(8)
arr
```

```python
# NumPy arrays can be sorted in-place using the sort method:
arr.sort()
arr
```

```python
arr = np.random.randn(5, 3)
arr
```

```python
# sort along the given axis
arr.sort(1)
arr
```

```python
# sort along the given axis
arr.sort(0)
arr
```

```python
# The top level method np.sort returns a sorted copy of an array instead of modifying
# the array in place.

arr = np.random.randn(8)
arr_sorted = np.sort(arr)

print(arr)
print()
print(arr_sorted)
```

Note:  The difference between .sort() and np.sort() is that .sort() change the original array while np.sort() create a sorted copy instead of modifying the original array

NumPy is able to save and load data to and from disk either in text or binary format.

## Storing Arrays on Disk in Binary Format

np.save and np.load are the two workhorse functions for efficiently saving and loading array data on disk. Arrays are saved by default in an uncompressed raw binary format with file extension .npy.

```python
arr = np.arange(10)
np.save('some_array', arr)
```

You will see a some_array.npy file in the directory on the left hand panel.

```python
arr_load = np.load('some_array.npy')
arr_load
```

# Linear Algebra

Linear algebra, like matrix multiplication, decompositions, determinants, and other square matrix math, is an important part of any array library.

Unlike some languages like MATLAB, multiplying two two-dimensional arrays with * is an element-wise
product instead of a matrix dot product.

As such, there is a function dot, both an array
method, and a function in the numpy namespace, for matrix multiplication:

```python
x = np.array([[1., 2., 3.], [4., 5., 6.]])
y = np.array([[6., 23.], [-1, 7], [8, 9]])

print(x)
print()
print(y)
```

```python
x.dot(y)
```

```python
np.dot(x, y)
```

```python
np.dot(x, np.ones(3))
```

numpy.linalg has a standard set of matrix decompositions and things like inverse and determinant

*   diag - Return the diagonal (or off-diagonal) elements of a square matrix as a 1D array, or convert a 1D array into a square matrix with zeros on the off-diagonal
*   dot  - Matrix multiplication
*   trace - Compute the sum of the diagonal elements
*   det - Compute the matrix determinant
*   eig - Compute the eigenvalues and eigenvectors of a square matrix
*   inv  - Compute the inverse of a square matrix
*   pinv - Compute the Moore-Penrose pseudo-inverse inverse of a square matrix
*   qr - Compute the QR decomposition
*   svd - Compute the singular value decomposition (SVD)
*   solve - Solve the linear system Ax = b for x, where A is a square matrix
*   lstsq - Compute the least-squares solution to y = Xb



```python
# numpy.linalg has a standard set of matrix decompositions and things like inverse and
# determinant

X = np.random.randn(5, 5)

mat = X.T.dot(X)
mat
```

```python
np.diag(mat)
```

```python
np.trace(mat)
```

```python
from numpy.linalg import det
det(mat)
```

```python
from numpy.linalg import eig

eig(mat)
```

```python
from numpy.linalg import inv

inv(mat)
```

## Random Number Generation

The numpy.random module supplements the built-in Python random with functions for efficiently generating whole arrays of sample values from many kinds of probability distributions.

Partial list of numpy.random functions
*  seed - Seed the random number generator
*  permutation - Return a random permutation of a sequence, or return a permuted range
*  shuffle - Randomly permute a sequence in place
*  rand - Draw samples from a uniform distribution
*  randint - Draw random integers from a given low-to-high range
*  randn - Draw samples from a normal distribution with mean 0 and standard deviation 1
*  binomial - Draw samples a binomial distribution
*  normal - Draw samples from a normal (Gaussian) distribution
*  beta - Draw samples from a beta distribution
*  chisquare - Draw samples from a chi-square distribution
*  gamma - Draw samples from a gamma distribution
*  uniform - Draw samples from a uniform [0, 1) distribution

```python
# For example, you can get a 4 by 4 array of samples from the standard
# normal distribution using normal:

samples = np.random.normal(size=(4, 4))
samples
```

Every time you run the random number generator, it will give you different number.

In order to ensure the same values are generated, set the seed number

```python
# run multiple time - you get the same values.
# if you remove np.random.seed(),  you get different values everytime you repeat.

np.random.seed(45)
samples = np.random.normal(size=(4, 4))
samples
```
